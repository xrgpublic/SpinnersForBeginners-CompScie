#ANSWER:

```python
import numpy as np


def prob(amplitude: complex) -> float:
    """Measurement probability  |⟨ψ|φ⟩|²  given an amplitude."""
    return float(np.abs(amplitude)**2)


# ---------- canonical basis vectors -----------------------------------------
# Using 1-D arrays as kets; NumPy treats them like column vectors in vdot/@
plus_z  = np.array([1, 0])        # |+z⟩
minus_z = np.array([0, 1])        # |-z⟩

"""
|+x⟩ = (1/√2) |+z⟩ + (1/√2) |-z⟩
|-x⟩ = (1/√2) |+z⟩ - (1/√2) |-z⟩
"""
# X-basis expressed in the Z-basis
plus_x  = (plus_z + minus_z) / np.sqrt(2)   # |+x⟩
minus_x = (plus_z - minus_z) / np.sqrt(2)   # |-x⟩

"""
|+y⟩ = (1/√2) |+z⟩ + (i/√2) |-z⟩
|-y⟩ = (1/√2) |+z⟩ - (i/√2) |-z⟩
"""
# Y-basis expressed in the Z-basis
plus_y  = (plus_z + minus_z * 1j) / np.sqrt(2)   # |+y⟩
minus_y = (plus_z - minus_z * 1j) / np.sqrt(2)   # |-y⟩

# Showing all the different computations:
# Remeber that the amplitude is found by np.vdot(bra, ket) which is the same as: ⟨ψ|φ⟩
# and the probability is found by np.abs(amplitude)**2); which is the same as |⟨ψ|φ⟩|²
tests = [
    ("|⟨+z|+x⟩|²",  np.abs(np.vdot(plus_z,  plus_x))**2),
    ("|⟨-z|+x⟩|²",  np.abs(np.vdot(minus_z, plus_x))**2),
    ("|⟨+z|-x⟩|²",  np.abs(np.vdot(plus_z,  minus_x))**2),
    ("|⟨-z|-x⟩|²",  np.abs(np.vdot(minus_z, minus_x))**2),
    ("|⟨+x|-x⟩|²",  np.abs(np.vdot(plus_x,  minus_x))**2),
    ("|⟨+x|+x⟩|²",  np.abs(np.vdot(plus_x,  plus_x))**2),
    ("|⟨+x|+y⟩|²",  np.abs(np.vdot(plus_x,  plus_y))**2),
    ("|⟨-x|+y⟩|²",  np.abs(np.vdot(minus_x, plus_y))**2),
    ("|⟨+x|-y⟩|²",  np.abs(np.vdot(plus_x,  minus_y))**2),
    ("|⟨-x|-y⟩|²",  np.abs(np.vdot(minus_x, minus_y))**2),

]

for label, prob in tests:
    print(f"{label:7s} probability = {prob:.3f}")
```
