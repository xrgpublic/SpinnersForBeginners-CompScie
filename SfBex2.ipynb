If you "take a step back" and look at the overall direction of travel, you will see that they go in the same direction as the plotted answers on EigenChris' answersheet.

```python
import matplotlib.pyplot as plt
import numpy as np

# Writing questions ( 1j is imaginary number i in python. 2j would be 2i, 1j^2 would be i^2 )
a = np.array([1, 1/2])
b = np.array([2, 1])
c = np.array([1j, 1])
d = np.array([-1, 1j])
e = np.array([1, np.exp(1j*(np.pi/4))])
f = np.array([np.exp(1j*(np.pi/4)), np.exp(1j*(np.pi/2))])

# Putting all questions into single array so we can iterate through it.
q = [a, b, c, d, e, f]

# Function to compute direction of jones vector
def jones(vec):
    amp = np.abs(vec) # Amplitude can be found by getting abs value.
    phi = np.angle(vec) # Functionally, this will take the imaginary part and return a pi/n equivalent
    t = np.linspace(0, np.pi*2, 50) # Creating an even amount of points between 0 and pi*2 (one full rotation around a circle)
    x = amp[0] * np.cos(t + phi[0]) # Calculating x component
    y = amp[1] * np.cos(t + phi[1]) # Calculating y component
    return x, y

# Goes through each question, gets its jones vector, then plots it.
for v in vectors:
    fig, ax = plt.subplots(figsize=(6, 6))
    x, y = jones(v)

    # Plotting points via scatter plot
    ax.scatter(x, y, s=10, color="blue", label="Path")

    # Plotting directional arrows on each scatter plot point
    dx = np.diff(x)
    dy = np.diff(y)
    ax.quiver(x[:-1], y[:-1], dx, dy, scale_units='xy', angles='xy', scale=1, width=0.005, color='red')

    ax.set_aspect('equal')
    ax.set_title(f"Jones vector: [{v[0]:.2f}, {v[1]:.2f}]")
    plt.grid(True)
    plt.legend()
    plt.show()
```
